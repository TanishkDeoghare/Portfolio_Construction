```python
import pandas as pd
import numpy as np

# Define paths to input files
filtered_spreads_stress_loss_path = r"C:/Users/tanishk.deoghare/OneDrive - Angel Oak Capital Advisors/Desktop/Portfolio Construction/Dry Run 2/Filtered_Mapped_Spreads_And_Stress_Loss_Report.xlsx"

# Load the filtered spreads and stress loss report
df_filtered_spreads_stress_loss = pd.read_excel(filtered_spreads_stress_loss_path)

# Define the sector allocation constraints
sector_constraints = {
    'ANGLX': {
        'ABS': (10, 30),
        'CLO': (0, 20),
        'Agency CMBS': (0, 15),
        'CMBS': (0, 5),
        'Agency RMBS': (10, 30),
        'Non-agency RMBS': (30, 60),
        'Credit': (5, 30),
    },
    'MBS': {
        'ABS': (0, 0),
        'CLO': (0, 0),
        'Agency CMBS': (0, 10),
        'CMBS': (0, 1),
        'Agency RMBS': (20, 60),
        'Non-agency RMBS': (40, 80),
        'Credit': (0, 0),
    },
    'AOUIX': {
        'ABS': (30, 55),
        'CLO': (0, 25),
        'Agency CMBS': (0, 10),
        'CMBS': (0, 0),
        'Agency RMBS': (10, 30),
        'Non-agency RMBS': (5, 30),
        'Credit': (5, 30),
    },
    'ASCIX': {
        'ABS': (0, 50),
        'CLO': (0, 30),
        'Agency CMBS': (0, 30),
        'CMBS': (0, 30),
        'Agency RMBS': (0, 30),
        'Non-agency RMBS': (0, 100),
        'Credit': (0, 30),
    }
}

# Define the credit rating constraints
credit_constraints = {
    'ANGLX': {
        'A': (0, 100),
        'AA': (0, 100),
        'AAA': (20, 100),
        'BBB': (0, 100),
        'BBB-': (0, 100),
        'BB': (0, 100),
        'B': (0, 100),
        'NR': (0, 100),
        'Below IG': (30, 70),  # Below IG includes BB, B, NR
    },
    'MBS': {
        'A': (0, 100),
        'AA': (0, 100),
        'AAA': (40, 100),
        'BBB': (0, 100),
        'BBB-': (0, 100),
        'BB': (0, 100),
        'B': (0, 100),
        'NR': (0, 10),
        'Below IG': (0, 10),  # Below IG includes BB, B, NR
    },
    'AOUIX': {
        'A': (0, 100),
        'AA': (0, 100),
        'AAA': (40, 100),
        'BBB': (0, 100),
        'BBB-': (0, 100),
        'BB': (0, 100),
        'B': (0, 100),
        'NR': (0, 5),
        'Below IG': (0, 5),  # Below IG includes BB, B, NR
    },
    'ASCIX': {
        'A': (0, 100),
        'AA': (0, 100),
        'AAA': (0, 100),
        'BBB': (0, 100),
        'BBB-': (0, 100),
        'BB': (0, 100),
        'B': (0, 100),
        'NR': (0, 100),
        'Below IG': (0, 100),  # Below IG includes BB, B, NR
    }
}

# Generate random numbers
num_random_numbers = 200
np.random.seed(42)  # Set a seed for reproducibility
random_numbers = np.random.random((len(df_filtered_spreads_stress_loss), num_random_numbers))

# Create a new DataFrame with instrument names and random numbers
random_numbers_df = pd.DataFrame(random_numbers, columns=[f'Random_Number_{i+1}' for i in range(num_random_numbers)])
df_random_numbers = pd.concat([df_filtered_spreads_stress_loss[['Instrument', 'Sector', 'Ratings']], random_numbers_df], axis=1)

# Normalize the numeric values and enforce constraints
def enforce_constraints(normalized_values, sector_constraints, credit_constraints):
    for alloc in sector_constraints.keys():
        # Ensure minimum sector constraints are met
        for sector, (min_val, max_val) in sector_constraints[alloc].items():
            normalized_values[alloc][sector] = max(normalized_values[alloc].get(sector, 0), min_val / 100.0)
        
        # Normalize sectors to sum to 1 after enforcing minimum constraints
        total = sum(normalized_values[alloc][sector] for sector in sector_constraints[alloc])
        if total == 0:
            total = 1  # To avoid division by zero
        normalized_values[alloc] = {sector: val / total for sector, val in normalized_values[alloc].items() if sector in sector_constraints[alloc]}
        
        # Ensure maximum sector constraints are not exceeded
        for sector, (min_val, max_val) in sector_constraints[alloc].items():
            if normalized_values[alloc][sector] > max_val / 100.0:
                excess = normalized_values[alloc][sector] - (max_val / 100.0)
                normalized_values[alloc][sector] = max_val / 100.0
                total_excess = sum(normalized_values[alloc][sector] for sector in sector_constraints[alloc]) - normalized_values[alloc][sector]
                for other_sector in normalized_values[alloc]:
                    if other_sector != sector and total_excess > 0:
                        proportion = normalized_values[alloc][other_sector] / total_excess
                        normalized_values[alloc][other_sector] += proportion * excess
        
        # Ensure minimum credit constraints are met
        for rating, (min_val, max_val) in credit_constraints[alloc].items():
            normalized_values[alloc][rating] = max(normalized_values[alloc].get(rating, 0), min_val / 100.0)
        
        # Normalize ratings to sum to 1 after enforcing minimum constraints
        total = sum(normalized_values[alloc][rating] for rating in credit_constraints[alloc])
        if total == 0:
            total = 1  # To avoid division by zero
        normalized_values[alloc].update({rating: val / total for rating, val in normalized_values[alloc].items() if rating in credit_constraints[alloc]})
        
        # Ensure maximum credit constraints are not exceeded
        for rating, (min_val, max_val) in credit_constraints[alloc].items():
            if normalized_values[alloc][rating] > max_val / 100.0:
                excess = normalized_values[alloc][rating] - (max_val / 100.0)
                normalized_values[alloc][rating] = max_val / 100.0
                total_excess = sum(normalized_values[alloc][rating] for rating in credit_constraints[alloc]) - normalized_values[alloc][rating]
                for other_rating in normalized_values[alloc]:
                    if other_rating != rating and total_excess > 0:
                        proportion = normalized_values[alloc][other_rating] / total_excess
                        normalized_values[alloc][other_rating] += proportion * excess
        
        # Handle Below IG constraint
        if 'Below IG' in credit_constraints[alloc]:
            below_ig_min, below_ig_max = credit_constraints[alloc]['Below IG']
            below_ig_sum = sum([normalized_values[alloc].get(rating, 0) for rating in ['BB', 'B', 'NR']])
            below_ig_min /= 100.0
            below_ig_max /= 100.0

            if below_ig_sum < below_ig_min:
                shortfall = below_ig_min - below_ig_sum
                for rating in ['BB', 'B', 'NR']:
                    if normalized_values[alloc].get(rating, 0) < credit_constraints[alloc].get(rating, (0, 100))[1] / 100.0:
                        increase = min(shortfall, (credit_constraints[alloc].get(rating, (0, 100))[1] / 100.0) - normalized_values[alloc].get(rating, 0))
                        normalized_values[alloc][rating] = normalized_values[alloc].get(rating, 0) + increase
                        shortfall -= increase
                        if shortfall <= 0:
                            break

            elif below_ig_sum > below_ig_max:
                excess = below_ig_sum - below_ig_max
                for rating in ['BB', 'B', 'NR']:
                    if normalized_values[alloc].get(rating, 0) > credit_constraints[alloc].get(rating, (0, 100))[0] / 100.0:
                        decrease = min(excess, normalized_values[alloc].get(rating, 0) - (credit_constraints[alloc].get(rating, (0, 100))[0] / 100.0))
                        normalized_values[alloc][rating] = normalized_values[alloc].get(rating, 0) - decrease
                        excess -= decrease
                        if excess <= 0:
                            break

        if 'AAA' in credit_constraints[alloc]:
            min_aaa, max_aaa = credit_constraints[alloc]['AAA']
            normalized_values[alloc]['AAA'] = min(max(normalized_values[alloc].get('AAA', 0), min_aaa / 100.0), max_aaa / 100.0)

        rating_total = sum([normalized_values[alloc].get(rating, 0) for rating in ['A', 'AA', 'AAA', 'BBB', 'BBB-', 'BB', 'B', 'NR']])
        if rating_total == 0:
            rating_total = 1  # To avoid division by zero
        for rating in ['A', 'AA', 'AAA', 'BBB', 'BBB-', 'BB', 'B', 'NR']:
            normalized_values[alloc][rating] = normalized_values[alloc].get(rating, 0) / rating_total

        normalized_values[alloc]['Below IG'] = sum([normalized_values[alloc].get(rating, 0) for rating in ['BB', 'B', 'NR']])

    return normalized_values

# Separate numeric columns for normalization
numeric_columns = [col for col in df_random_numbers.columns if col.startswith('Random_Number_')]
df_numeric = df_random_numbers[numeric_columns]

# Normalize the numeric values
normalized_df = df_numeric.div(df_numeric.sum(axis=1), axis=0)

# Add back the non-numeric columns
df_normalized = pd.concat([df_random_numbers[['Instrument', 'Sector', 'Ratings']], normalized_df], axis=1)

# Initialize a dictionary to store normalized values for each allocation
normalized_values = {alloc: df_normalized[df_normalized['Sector'] == alloc].set_index(['Sector', 'Ratings']).to_dict(orient='index') for alloc in sector_constraints.keys()}

# Enforce constraints on the normalized values
constrained_values = enforce_constraints(normalized_values, sector_constraints, credit_constraints)

# Convert the constrained values dictionary back to a DataFrame for saving or further processing
constrained_df = pd.DataFrame(constrained_values)

# Save the constrained DataFrame to an Excel file (optional)
constrained_df.to_excel('Constrained_Values.xlsx', index=True)

# Output the constrained DataFrame for verification
print(constrained_df.head())
```
